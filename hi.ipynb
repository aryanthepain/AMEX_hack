```python
null_counts = train_df.isnull().sum()
total_rows = len(train_df)
```

```python
high_null_cols = null_counts[null_counts > 0.6 * total_rows].index.tolist()
moderate_null_cols = null_counts[(null_counts > 0.1 * total_rows) & (null_counts <= 0.6 * total_rows)].index.tolist()
low_null_cols = null_counts[(null_counts > 0) & (null_counts <= 0.1 * total_rows)].index.tolist()
```

```python
data_dict['Description'] = data_dict['Description'].astype(str).str.strip()
data_dict['Type'] = data_dict['Type'].astype(str).str.strip().str.lower()
print('types changed')

categorical_cols = data_dict[data_dict['Type'].str.contains('categorical')]['Description'].tolist()
print('categorical_cols extracted')
one_hot_cols = data_dict[data_dict['Type'].str.contains('one hot')]['Description'].tolist()
print('one_hot_cols extracted')
numerical_cols = data_dict[data_dict['Type'].str.contains('numerical')]['Description'].tolist()
print('numerical_cols extracted')

categorical_cols = [col for col in categorical_cols if col in train_df[target_columns].columns]
print('categorical_cols changed')
one_hot_cols = [col for col in one_hot_cols if col in train_df[target_columns].columns]
print('one_hot_cols changed')
numerical_cols = [col for col in numerical_cols if col in train_df[target_columns].columns]
print('numerical_cols changed')

# Output counts for verification
len(categorical_cols), len(one_hot_cols), len(numerical_cols)
```

```python
# High-missing
print(len(high_null_cols))
i=0
for col in high_null_cols:
    print(i, col)
    if col in numerical_cols:
        train_df[col] = train_df[col].fillna(0)
    elif col in one_hot_cols:
        train_df[col] = train_df[col].fillna(0)
    elif col in categorical_cols:
        train_df[col] = train_df[col].fillna('missing')
    i+=1

# Moderate-missing
i=0
for col in moderate_null_cols:
    print(i, col)
    if col in numerical_cols:
        train_df[col] = train_df[col].fillna(train_df[col].median())
    elif col in one_hot_cols:
        train_df[col] = train_df[col].fillna(0)
    elif col in categorical_cols:
        train_df[col] = train_df[col].fillna(train_df[col].mode()[0])
    i+=1

# Low-missing
i=0
for col in low_null_cols:
    print(i, col)
    if col in numerical_cols:
        train_df[col] = train_df[col].fillna(train_df[col].median())
    elif col in one_hot_cols:
        train_df[col] = train_df[col].fillna(0)
    elif col in categorical_cols:
        train_df[col] = train_df[col].fillna(train_df[col].mode()[0])
    i+=1
```

# AMEX Hackathon: Preprocessing Review & Recommendations

This notebook reviews the preprocessing pipeline, analyzes the data, and provides recommendations for optimal feature engineering and dimensionality reduction.

```python
# Import libraries and load data
import pandas as pd
import numpy as np

# Load the first 500 rows of train and test data
train_df = pd.read_csv('output/train_df_head.csv')
test_df = pd.read_csv('output/test_df_head.csv')

print(f"Train shape: {train_df.shape}")
print(f"Test shape: {test_df.shape}")
```

## 1. Target Variable & Class Distribution

The target variable is `y` (1 = click, 0 = no click). Below, we analyze the class distribution to check for imbalance.

```python
# Class distribution of target variable
y_counts = train_df['y'].value_counts()
y_ratio = train_df['y'].value_counts(normalize=True)
print('Class counts:')
print(y_counts)
print('\nClass ratio:')
print(y_ratio)
```

## 2. ID Columns

The dataset contains 5 ID columns (`id1` to `id5`). These are not used as features, but are important for grouping, stratification, and MAP@7 calculation.

## 3. Missing Value Analysis

We analyze missing values and review the imputation strategy. Features with high missingness may be dropped or filled with 0; moderate/low missingness can be imputed with median/mode.

```python
# Missing value analysis
missing_counts = train_df.isnull().sum()
missing_ratio = missing_counts / len(train_df)
missing_summary = pd.DataFrame({'missing_count': missing_counts, 'missing_ratio': missing_ratio})
missing_summary = missing_summary[missing_summary['missing_count'] > 0].sort_values('missing_ratio', ascending=False)
missing_summary.head(20)
```

## 4. Categorical Features & Cardinality

We check for categorical features and their cardinality. High-cardinality features should be encoded with target/frequency encoding or embeddings, not one-hot.

```python
# Categorical features and cardinality
cat_cols = train_df.select_dtypes(include='object').columns
for col in cat_cols:
    print(f"{col}: {train_df[col].nunique()} unique values")
```

## 5. Feature Engineering

- Extract temporal features (day, hour, recency) from date/time columns.
- Aggregate statistics by user/offer (mean, count, etc.).
- Consider interaction features (user-offer, user-time).

## 6. Dimensionality Reduction: PCA & Feature Selection

- Use PCA only on standardized numerical features if overfitting or multicollinearity is observed.
- For tree-based models, prefer feature selection (e.g., model-based, mutual information) over PCA.
- Drop features with extreme missingness if not useful.

```python
# Example: PCA on numerical features
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Select numerical columns (excluding IDs and target)
numeric_cols = train_df.select_dtypes(include=[np.number]).columns.difference(['y'])
X_num = train_df[numeric_cols].fillna(0)

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_num)

# Apply PCA (keep 95% variance)
pca = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X_scaled)
print(f"Original num features: {X_num.shape[1]}, PCA features: {X_pca.shape[1]}")
```

## 7. Evaluation Metric & Validation Strategy

- Metric: MAP@7 (Mean Average Precision at 7) for ranking offers per user.
- Plan for user-based or time-based validation to avoid leakage and mimic test conditions.

## 8. Recommendations & Next Steps

- Use advanced imputation for moderate missingness if time allows.
- Engineer time-based and aggregated features.
- Use PCA only if justified; otherwise, prefer feature selection.
- Prepare for MAP@7 evaluation by grouping and ranking predictions per user.
- Document all steps and rationale for reproducibility.
